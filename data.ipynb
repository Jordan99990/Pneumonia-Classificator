```python
from silence_tensorflow import silence_tensorflow # just for not showing the warning message -_-
silence_tensorflow("ERROR")

import os
import keras
from tensorflow.keras.preprocessing.image import load_img, img_to_array # type: ignore
import matplotlib.pyplot as plt
import numpy as np
```

```python
data_normal = 'Dataset/NORMAL'
data_pneumonia = 'Dataset/PNEUMONIA'
```

```python
def load_and_display_images(directory, label, num_images=5):
    files = os.listdir(directory)
    plt.figure(figsize=(15, 5))
    for i, file in enumerate(files[:num_images]):
        img_path = os.path.join(directory, file)
        img = load_img(img_path)
        img_array = img_to_array(img)
        plt.subplot(1, num_images, i + 1)
        plt.imshow(img_array.astype('uint8')) 
        plt.title(f'{label} {i+1}')
        plt.axis('off')
    plt.show()
```

```python
print("Normal Images:")
load_and_display_images(data_normal, 'Normal')
```

```python
print("Pneumonia Images:")
load_and_display_images(data_pneumonia, 'Pneumonia')
```

```python
num_normal_images = len(os.listdir(data_normal))
num_pneumonia_images = len(os.listdir(data_pneumonia))

print(f'Number of Normal images: {num_normal_images}')
print(f'Number of Pneumonia images: {num_pneumonia_images}')
```
